```python
#Menggabungkan data 'bekasi' dan 'bogor
dfx = pd.concat([dfB, dfb])
dfx
```

```python
dfx.info()
```

```python
#Mencari IQR untuk menghilangkan outliers
Q1 = dfx.quantile(0.25)
Q3 = dfx.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
dataF = dfx[~((dfx<(Q1-1.5*IQR))|(dfx>(Q3+1.5*IQR))).any(axis=1)]
dataF.shape, dfx.shape
#dataF merupakan data yang sudah dihilangkan outliernya
#dfx merupakan data sebelum dihilangkan outliernya
```

* 35 - 25 = 10
* Sebanyakan 10 data yang merupakan outlier telah dibuang dari data menggunakan metode IQR
